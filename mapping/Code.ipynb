```python
fg=folium.FeatureGroup(name="My Map")
```

```python
for coordinates in [[50, -100], [45, -100]]:
        fg.add_child(folium.Marker(location=coordinates, popup= "Here it is!", icon=folium.Icon(color="green")))
```

```python
#reading locations from a file
```

```python
data=pd.read_csv("Volcanoes.txt")
```

```python
data
```

```python
lat=list(data["LAT"])
```

```python
lon=list(data["LON"])
```

```python
for lt, ln in zip(lat,lon):
    fg.add_child(folium.Marker(location=[lt, ln], popup= "Here it is!", icon=folium.Icon(color="green")))
```

```python
map.add_child(fg)
```

```python
#adding more information to the popup window
```

```python
elev=list(data["ELEV"])
```

```python
for lt, ln, el in zip(lat,lon, elev):
    fg.add_child(folium.Marker(location=[lt, ln], popup=str(el) , icon=folium.Icon(color="green")))
```

```python
map.add_child(fg)
```

```python
map.save("Map_1.html")
```
